# Feature Selection/Vector Assembler

```python
df = DF_New.select('sport','dsport','sbytes','dbytes','sttl','dttl','sloss','dloss','ct_dst_src_ltm','is_ftp_login','Label')
cols=df.columns
cols.remove("Label")
from pyspark.ml.feature import VectorAssembler
VA_Data = VectorAssembler(inputCols=cols,outputCol="features")
#Features is the name of the column which will contain the feature vector.

DataFrame_BC=VA_Data.transform(DF_New)
DataFrame_BC.show(3)
DataFrame_BC.select("features").show(truncate=False)
```

```python
# now apply the normalization of the data in the dataframe , 
#we will input thecolumn that was named as "features" in the last step and 
#after standarziationoperation the output column will be naed as "Scaled Features"
#Standard Sclarizer
from pyspark.ml.feature import StandardScaler
ScaledData=StandardScaler().setInputCol("features").setOutputCol("Sc_features")
DataFrame_BC=ScaledData.fit(DataFrame_BC).transform(DataFrame_BC)
DataFrame_BC.select("features","Sc_features").show(5) 
# showing the difference in the features and scaled features
```

# Feature Selection

```python
#get correlation matrix
'''
Correlation calculates the correlation between 2 column of dataframe as a double value
It measures the way 2 var changes with respect to each other,
and Provides quantative measurement of static dependence between 2 random varoable
'''
from pyspark.ml.stat import Correlation
Datacorr = Correlation.corr(DataFrame_BC, "features")
Datacorr.collect()[0]["pearson({})".format("features")].values  
# below sparse vector gives correlation between the vector column DataFrame and the output column features
 
```

# Data Split

```python
# Now after the addition of the columns for standarization and feature vectorin the Data Frame we will divided this Data Frame as train and Test
#Train, test split
# generate random column with independent and identically distributed sample
train_lr, test_lr  = DataFrame_BC.randomSplit([0.8,0.2],seed =12345)
print("Logistic Regression Train Data total: " + str(train_lr.count()))
print("Logistic Regression Test Data total: " + str(test_lr.count()))
```

# Data Balancing

```python
#Before we use the training data set to train the algorthim we will 
#check if it balanced or imbalanced data. Not we are only checking on training set
# Read the total count of the class labels
BCDataSize=float(train_lr.select("Label").count())
PositiveNos=train_lr.select("Label").where('Label == 1').count()
Each_one=(float(PositiveNos)/float(BCDataSize))*100
NegativeNos=float(BCDataSize-PositiveNos)
print('count of ones {}'.format(PositiveNos))
print('Percentage of ones {}'.format(Each_one))
print("BC Total Data count",BCDataSize, "Postive Records",PositiveNos,"Negative records",NegativeNos)
```

```python
BalanceRatio= NegativeNos/BCDataSize
print('Ratio of balanced data = {}'.format(BalanceRatio))
```

```python
# Here we adjust the weights for the training data to tackle class imbalance, 
#it will add one more column in the training data and we named it as "Weight_Class"
from pyspark.sql.functions import when
train_lr=train_lr.withColumn("Weight_Class",when(train_lr.Label == 1,BalanceRatio).otherwise(1-BalanceRatio))
train_lr.select("Weight_Class","Label").show(25)
```

# Binary Classification Model 

```python
'''
logistic regression we predict classes. 
Independent variable x and dependentvariable y. 
Does not give continous value.
->Sigmoid cost function has two value 0 and 1 
1. > 0.5 Sigmoid classify as 1 
2. < 0.5 Sigmoid classify as 0
w-weight set during iteration during performance
In our data set we need to classify wether attack is normal or not.
'''
```

```python
from pyspark.ml.classification import LogisticRegression
lrBC = LogisticRegression(labelCol="Label", featuresCol="features",maxIter=10,weightCol="Weight_Class")
model_lrBC=lrBC.fit(train_lr)
train_lr=model_lrBC.transform(train_lr)
test_lr=model_lrBC.transform(test_lr)
test_lr.select("Label","features").show(10)
#fit - is a method it takes the original or raw data and labels from our dataset. 
#transform - takes original data and labels and joins it together
 
```

# Data Evaluation

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
BC_Evaluation=BinaryClassificationEvaluator(rawPredictionCol="rawPrediction",labelCol="Label")
test_lr.select("Label","rawPrediction","prediction","probability").show(5)
print("Train data's area under ROC is {}".format(BC_Evaluation.evaluate(train_lr)))
print("Test data's area under ROC is {}".format(BC_Evaluation.evaluate(test_lr)))


#evaluator- evaluvates the model performance
```

```python
accuracyLRTrain = BC_Evaluation.evaluate(train_lr)
accuracyLRTest = BC_Evaluation.evaluate(test_lr)
print('Train Accuracy for Binary Classification =',accuracyLRTrain)
print('Test Accuracy for Binary Classification =',accuracyLRTest)
```

# Feature Extraction

```python
'''
Extracted the selected features and passed it to one data frame df_mc.
and displayed its data type using printSchema()
'''
df_mc = DF_New.select('sport','dur','sbytes','dbytes','sttl','dttl','sloss','dloss','Sload','Dload','Sintpkt','Dintpkt','ackdat','is_ftp_login','attack_cat')
cols = df_mc.columns
df_mc.printSchema()
```

```python
'''
From the datatypes listed above, we selected attack_cat column which is of string type for further processing
to understand about the class in it, we have grouped the class in it by useing groupBy()
and we listed the total count of data present in each class.
'''

df_mc.groupby('attack_cat').count().show()
```

# Data Cleaning

```python
'''
1. Merging Class Backdoors,Backdoor and Reconnaissance,Reconnaissance  
   to one class as Backdoor and Reconnaissance respectively.
2. as we had 2218456 null values in null class, to use it further in our classification,
   we filled the null values as normal using fillna()
'''
```

```python

from pyspark.sql.functions import when
df_mc=df_mc.withColumn("attack_cat",when(df_mc.attack_cat=='Backdoors','Backdoor').otherwise(df_mc.attack_cat))
df_mc=df_mc.withColumn("attack_cat",when(df_mc.attack_cat=='Reconnaissance ','Reconnaissance').otherwise(df_mc.attack_cat))
df_mc=df_mc.fillna('Normal')
df_mc.groupby('attack_cat').count().sort("count", ascending=True).show()
```

# Vector Assembler

```python
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
MC_Vector = VectorAssembler(inputCols=['sport','dur','sbytes','dbytes','sttl','dttl','sloss','dloss','Sload','Dload','Sintpkt','Dintpkt','ackdat','is_ftp_login'],outputCol="featuresMul")
VectorData = MC_Vector.transform(df_mc)
VectorData.show(3)
```

```python
#Letâ€™s remove unnecessary columns:
MCFeature = VectorData.drop('sport','dur','sbytes','dbytes','sttl','dttl','sloss','dloss','Sload','Dload','Sintpkt','Dintpkt','ackdat','is_ftp_login')
MCFeature.show(10)
```

```python
from pyspark.ml.feature import StringIndexer
MCSTR_INT = StringIndexer(inputCol="attack_cat", outputCol="attack_catIndex")
MCFeature = MCSTR_INT.fit(MCFeature).transform(MCFeature)
MCFeature.select('attack_cat','attack_catIndex','featuresMul').distinct().show(10)
```

# Data Split

```python
(DTTrainData, DTTestData) = MCFeature.randomSplit([0.7, 0.3])
print("Decision Tree Train Data total: " + str(DTTrainData.count()))
print("Decision Tree Test Data total: " + str(DTTestData.count()))
```

```python
from pyspark.ml.classification import DecisionTreeClassifier

DTMC = DecisionTreeClassifier(labelCol="attack_catIndex", featuresCol="featuresMul")
DTMC_M= DTMC.fit(DTTrainData)
DTMCtrain= DTMC_M.transform(DTTrainData)
DTMCtest = DTMC_M.transform(DTTestData)
DTMCtest.select("attack_cat","featuresMul","prediction").show(10)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
DTMCEvaluation= MulticlassClassificationEvaluator(labelCol="attack_catIndex", predictionCol="prediction", metricName="accuracy")
DTMCTestAcc = DTMCEvaluation.evaluate(DTMCtest)
print("Test data accuracy for Decision tree = " , DTMCTestAcc)


DTMCTrainAcc = DTMCEvaluation.evaluate(DTMCtrain)
print("Train data accuracy for Decision tree = " , DTMCTrainAcc)
```

# Bagging

```python
#Bagging using randomforest
#The random forest is well known for bagging method
#here trees are fitted on the sample of bootstrap
# lastly, the fitted trees are combined and give better output with lower variance.

from pyspark.ml.classification import RandomForestClassifier
RFMC = RandomForestClassifier(featuresCol='featuresMul', labelCol='attack_catIndex', maxDepth = 4,numTrees=20)#Number of tree bagging
RFMCtraining, RFMCtesting = MCFeature.randomSplit([0.7, 0.3], seed = 2018)
RFMC_M = RFMC.fit(RFMCtraining)
RFMCtrain= RFMC_M.transform(RFMCtraining)
RFMCtest = RFMC_M.transform(RFMCtesting)

#predictions.show()
RFMCtest.select('attack_cat','rawPrediction','probability','prediction').show()


```

```python
RFMCEvaluation = MulticlassClassificationEvaluator(labelCol="attack_catIndex", predictionCol="prediction",metricName="accuracy")

RFMCTrainAcc = RFMCEvaluation.evaluate(RFMCtrain)
print("RandomForest Accuracy for training Dataset = " , RFMCTrainAcc)

RFMCTestAcc = RFMCEvaluation.evaluate(RFMCtest)
print("RandomForest Accuracy for testing Dataset  = " , RFMCTestAcc)
print(RFMC_M.toDebugString)
```
